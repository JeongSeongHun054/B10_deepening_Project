```python
import sklearn
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/취업준비/스파르타코딩/머신러닝(주피터노트북)/train.csv')
```

```python

type(df['timestamp'][0])
```

```python
import datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
type(df['timestamp'][0])
```

```python
# timestamp : year / month / day 쪼개기

df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['dayofweek'] = df['timestamp'].dt.dayofweek
df['ttimestampp'] = df['year']+df['month']## 무슨요일인지 확인 
df['Index'] = df.index


df
```

```python
Brocoly_data = df[df['item'] == 'BC']
Brocoly_data
```

```python
#Brocoly_data.plot()
#plt.show()
sns.scatterplot( data = Brocoly_data, x = 'supply(kg)' , y = 'price(원/kg)')
sns.lineplot( data = Brocoly_data, x = 'supply(kg)' , y = 'price(원/kg)')
#plt.figure(figsize=(40,20))
```

```python
sns.set(rc={"figure.figsize":(80, 10)}) #width=3, #height=4
sns.scatterplot( data = Brocoly_data, x = 'supply(kg)' , y = 'price(원/kg)')
sns.lineplot( data = Brocoly_data, x = 'supply(kg)' , y = 'price(원/kg)')
sns.regplot( data = Brocoly_data, x = 'supply(kg)' , y = 'price(원/kg)')
sns.lmplot( data = Brocoly_data, x = 'supply(kg)' , y = 'price(원/kg)')
```

```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
type(model_lr)
```

```python
prsu_df = Brocoly_data[['supply(kg)', 'price(원/kg)']]
model_lr.fit(Brocoly_data[['supply(kg)']], Brocoly_data['price(원/kg)'])
```

```python
y_pred = model_lr.predict(Brocoly_data[['supply(kg)']])
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Brocoly_data['price(원/kg)'], y_pred)
```

```python
from sklearn.metrics import mean_squared_error 
mean_squared_error(Brocoly_data['price(원/kg)'], y_pred)
```

```python
from sklearn.metrics import mean_squared_error 
MSE = mean_squared_error(Brocoly_data['price(원/kg)'], y_pred) 
MSE ** 0.5
```

```python
from sklearn.metrics import mean_squared_log_error
mean_squared_log_error(Brocoly_data['price(원/kg)'], y_pred)
```

```python
def MAPE (Brocoly_data['price(원/kg)'] , y_pred) :
	return np.mean(np.abs((Brocoly_data['price(원/kg)'] - y_pred) / Brocoly_data['price(원/kg)'])) * 100 
    
MAPE(Brocoly_data['price(원/kg)'], y_pred)
```

```python
def MAE(y_test, y_pred): 
	return np.mean((Brocoly_data['price(원/kg)'] - y_pred) / Brocoly_data['price(원/kg)']) * 100)
    
MAE(y_test, y_pred)
```
